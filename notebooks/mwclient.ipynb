# recursively get all subcats/pags for a given cat

```python
jsb_cat = mw.site.Categories["Johann_Sebastian_Bach"]
list(jsb_cat.members())


def category_leaves(catg, path=tuple(), max_depth=None):
    
    # check that max_path is not exceeded
    if max_depth is not None and len(path) >= max_depth:
        raise StopIteration
    
    # check whether member is already in the path -- to prevent infinite loops from cyclic graphs
    #if catg.name in set([c.name for c in path]):
    #    raise StopIteration
        # print ("cyclic loop", [c.name for c in path] + [catg.name])
    
    # otherwise go ahead
    for member in catg.members():
        if isinstance(member, mwclient.listing.Category):
            if not member.name in set([c.name for c in path]):
                for (child, child_path) in category_leaves(member, path + (member,), max_depth):
                    yield (child, child_path)
        elif isinstance(member, mwclient.page.Page):
            yield (member, path)
```

```python
from collections import defaultdict, Counter

pages = Counter()
categories = Counter()

for (i,(leaf, path)) in enumerate(islice(category_leaves(jsb_cat,max_depth=1000),None)):
    
    print ("\r {0}".format(i),  " > ".join( [p.name for p in path] + [leaf.name] ), end="")
    pages.update([leaf])
    if len(path):
        categories.update([path[-1].name])
    else:
        categories.update([None])
```

```python
# https://github.com/legoktm/wikidata/blob/67ef5590c4269d0caed2736dfcc640a9985e03cb/migrate_genders.py

from itertools import islice

import pywikibot
repo = pywikibot.Site('en','wikipedia').data_repository()
male = pywikibot.ItemPage(repo, 'Q6581097')
female = pywikibot.ItemPage(repo, 'Q6581072')

def gen():
    oldmale = pywikibot.ItemPage(repo, 'Q44148')
    oldfemale = pywikibot.ItemPage(repo, 'Q43445')
    for page in oldmale.getReferences(namespaces=[0]):
        yield page
    for page in oldfemale.getReferences(namespaces=[0]):
        yield page


list(islice(gen(),10))
```
